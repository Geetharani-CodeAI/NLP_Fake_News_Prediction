```python
eng_sentences = train["en"]
tamil_sentences = train["ta"]
eng_sentences = eng_sentences.head(1000)
tamil_sentences = tamil_sentences.head(1000)
tamil_sentences
```

```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
def sentToken(sentence):
    # defining the dataset
    dataset = sentence

    # convert each sentence to a list of words
    sentences = [sentence.split() for sentence in dataset]

    # Print the result
    print(sentences)

    return sentences
```

```python
engSentence = sentToken(eng_sentences)
print(engSentence)
```

```python
tamilSentence = sentToken(tamil_sentences)
print(tamilSentence)
```

```python
def ownWordModel(langsentence, modelname):
    model = Word2Vec(langsentence, min_count =1)
    print(model)
    # save model
    model.save(modelname)
    # fit a 2d PCA model to the vectors
    X = model.wv.vectors # Corrected line  # wv - Weight vector
    pca = PCA(n_components = 2)
    result = pca.fit_transform(X)

    # create a scatter plot of the projection
    plt.scatter(result[:, 0], result[:, 1])
    words = list(model.wv.index_to_key)
    for i, word in enumerate(words):
        plt.annotate(word, xy= (result[i,0], result[i, 1]))
    plt.show()
```

```python
ownWordModel(engSentence, 'engmodel.bin')
```

```python
ownWordModel(tamilSentence,"tamilmodel.bin")
```
