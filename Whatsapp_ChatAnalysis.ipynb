## Sentimental Analysis

```python
# Downloading vader_lexicon for the process
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
dataset = df
dataset.dropna(inplace = True)
```

```python
sid = SentimentIntensityAnalyzer()
senti = sid.polarity_scores(dataset['Chat'][4])
senti
```

```python
def sentimentAnalysis(data, columnname):
    import matplotlib.pyplot as plt
    import seaborn as sns

    # importing necessary Packages
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    sid = SentimentIntensityAnalyzer()

    # Deleting null values if any
    data.dropna(inplace = True)

    # checking
    #sid.polariy_scores(data[columnname][45])

    # Checking respective columns
    data['scores'] = data[columnname].apply(lambda commenText: sid.polarity_scores(commenText))
    data['compound'] = data['scores'].apply(lambda score_dict: score_dict['compound'])
    data['Negative'] = data['scores'].apply(lambda score_dict: score_dict['neg'])
    data['Positive'] = data['scores'].apply(lambda score_dict: score_dict['pos'])
    data['Neutral'] = data['scores'].apply(lambda score_dict: score_dict['neu'])

    # Compound Score Analysis
    data['Comp_Score'] = data['compound'].apply(lambda c:'Pos' if c>=0 else 'Neg')
    # plt.clf()

    # Checking how many Pos and Neg
    posneg = pd.DataFrame(data['Comp_Score'].value_counts())
    return posneg, data
```

```python
pos, data_Senti = sentimentAnalysis(dataset, columnname = 'Chat')
```

## Topic Modelling

#### TFIDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_df = 0.95, min_df = 2, stop_words = 'english')
dtm = tfidf.fit_transform(df['Chat'])

from sklearn.decomposition import NMF
nmf_model = NMF(n_components = 10, random_state = 42)
nmf_model.fit(dtm)

for index, topic in enumerate(nmf_model.components_):
    results = ([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(index,results)  

topic_results = nmf_model.transform(dtm)
topic_results
df["Topic"] = topic_results.argmax(axis = 1)   

df
```

#### TFIDF with LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_df = 0.95, min_df = 2, stop_words = 'english')
dtm = tfidf.fit_transform(df1['Chat'])

lda = LatentDirichletAllocation(n_components =6, random_state = 42) 
lda.fit(dtm)

for index, topic in enumerate(lda.components_):
    results = ([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(index,results)  

topic_results = lda.transform(dtm)
topic_results
df1["Topic"] = topic_results.argmax(axis = 1)   

df1
```

#### Counter Vectorizer with LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(stop_words = "english")
dtm = count_vectorizer.fit_transform(df1['Chat'])

lda = LatentDirichletAllocation(n_components = 5, random_state = 42) 
lda.fit(dtm)

for index, topic in enumerate(lda.components_):
    results = ([count_vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(index,results)  

topic_results = lda.transform(dtm)
topic_results
df1["Topic"] = topic_results.argmax(axis = 1)   

df1
```

## Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "".join(df['Chat'].astype(str))
wc = WordCloud(background_color = 'pink', stopwords = 'english', 
               max_words = 500).generate(text)

plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

#### Word Cloud for each topic

```python
for i in range(5):
    topic_words = " ".join(df[df['Topic']==i]["Chat"].astype(str))
    wc = WordCloud(background_color = 'white').generate(topic_words)
    plt.imshow(wc)
    plt.title(f"Topic {i}")
    plt.axis("off")
    plt.show()
```
